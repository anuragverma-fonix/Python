# Matplotlib and Seaborn Tutorial for Beginners

This notebook covers various plotting techniques using Matplotlib and Seaborn libraries in Python. Each section includes examples with comments to help beginners understand the concepts.

```python
import matplotlib.pyplot as plt
import seaborn as sns  # For advanced statistical plots
import numpy as np
import pandas as pd

# Set up the plotting style
plt.style.use('default')  # Use default style for consistency
sns.set_theme()  # Apply Seaborn's default theme
```

## 1. Line Plot

A line plot is used to display data points connected by straight lines. It's useful for showing trends over time or continuous data.

```python
x = [1, 2, 3, 4, 5]  # X-axis values (could be time periods)
y = [2, 4, 6, 8, 10]  # Y-axis values (data points)

# Create the line plot
plt.plot(x, y)  # Plot x against y with a line connecting the points

# Add labels and title
plt.xlabel('X Values')  # Label for x-axis
plt.ylabel('Y Values')  # Label for y-axis
plt.title('Simple Line Plot')  # Title of the plot

# Display the plot
plt.show()  # This command displays the plot
```

## 2. Bar Chart

A bar chart represents data with rectangular bars. The height of each bar corresponds to the value it represents.

```python
# Sample data for bar chart
categories = ['A', 'B', 'C', 'D']  # Categories for bars
values = [10, 20, 15, 25]  # Heights of the bars

# Create bar chart
plt.bar(categories, values)  # Each bar represents a category and its value

# Add labels and title
plt.xlabel('Categories')  # Label for x-axis
plt.ylabel('Values')  # Label for y-axis
plt.title('Bar Chart Example')  # Title

# Display the plot
plt.show()
```

## 3. Scatter Plot

A scatter plot displays points based on two variables. It's useful for showing relationships or correlations between variables.

```python
# Sample data for scatter plot
x = [1, 2, 3, 4, 5, 6]  # X coordinates
y = [2, 3, 5, 7, 11, 13]  # Y coordinates

# Create scatter plot
plt.scatter(x, y)  # Each point represents a pair (x, y)

# Labels
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.title('Scatter Plot Example')

# Show plot
plt.show()
```

## 4. Histogram

A histogram shows the distribution of a dataset by dividing it into bins and counting the number of observations in each bin.

```python
# Generate random data for histogram
data = np.random.normal(0, 1, 1000)  # 1000 random numbers from a normal distribution

# Create histogram
plt.hist(data, bins=50)  # 'bins' specifies how many bars; higher bins = more detail

# Labels
plt.xlabel('Value')
plt.ylabel('Frequency')  # How many data points fall into each bin
plt.title('Histogram Example')

# Show
plt.show()
```

## 5. Styling & Labeling

This section demonstrates how to customize plots with colors, line styles, legends, and better formatting for clarity.

```python
# Sample data

x = np.linspace(0, 10, 100)  # 100 points from 0 to 10
y1 = np.sin(x)  # Sine wave
y2 = np.cos(x)  # Cosine wave

# Plot with custom styling
plt.plot(x, y1, label='Sine Wave', color='blue', linestyle='-', linewidth=2, marker='o', markersize=3)
plt.plot(x, y2, label='Cosine Wave', color='red', linestyle='--', linewidth=2, marker='s', markersize=3)

# Enhanced labels and title
plt.title('Styled Line Plot with Markers', fontsize=16, fontweight='bold')
plt.xlabel('X Axis (Time)', fontsize=12)
plt.ylabel('Y Axis (Amplitude)', fontsize=12)

# Add legend
plt.legend(loc='upper right')  # Position the legend

# Add grid for better readability
plt.grid(True, alpha=0.5)  # Semi-transparent grid

plt.show()
```

## 6. Introduction to Seaborn

Seaborn is a Python data visualization library based on Matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.

## 7. Heatmaps

A heatmap is a graphical representation of data where values are depicted by color. It's useful for visualizing matrices or correlation data.

```python
# Create a sample 2D array (matrix)
data = np.random.rand(8, 8)  # 8x8 matrix with random values

# Create heatmap
sns.heatmap(data, annot=True, cmap='coolwarm', fmt='.2f')

# 'annot=True' shows the values in each cell
# 'cmap' chooses the color scheme
# 'fmt' formats the annotation numbers

plt.title('Heatmap Example')

plt.show()
```

## 8. Pair Plots

Pair plots visualize pairwise relationships in a dataset. For each pair of variables, it shows scatter plots, and for single variables, histograms or density plots.

```python
# Create a sample DataFrame with multiple variables

np.random.seed(42)  # For reproducible results

df = pd.DataFrame({
    'Variable_A': np.random.normal(0, 1, 100),  # Normal distribution
    'Variable_B': np.random.normal(2, 1.5, 100),  # Another normal
    'Variable_C': np.random.exponential(1, 100),  # Exponential distribution
    'Category': np.random.choice(['Group1', 'Group2'], 100)  # Categorical for hue
})

# Create pair plot
sns.pairplot(df, hue='Category', diag_kind='kde')

# 'hue' colors points by category
# 'diag_kind' specifies diagonal plots (kde for density)

plt.show()
```

## 9. Distribution Plots

Distribution plots show how a variable is distributed. They often combine histograms with kernel density estimates for a smooth representation.

```python
# Generate sample data

data = np.random.normal(0, 1, 1000)  # 1000 points from normal distribution

# Create distribution plot
sns.histplot(data, kde=True, bins=30, color='skyblue', alpha=0.7)

# 'kde=True' adds the kernel density estimate curve
# 'bins' controls histogram bars
# 'color' and 'alpha' for styling

plt.title('Distribution Plot with KDE')
plt.xlabel('Value')
plt.ylabel('Density')

plt.show()
```
